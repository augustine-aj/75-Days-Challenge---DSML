## Topics to Cover:
### 1. Introduction to `Matplotlib`:
Matplotlib is a comprehensive library used to create static, animated, and interactive visualizations in Python.
It supports various types of plots such as line plots, scatter plots, bar charts, histograms, and more.

---

### 2. Creating Simple Plots:

#### ***Line Plots***: 
Used for visualizing trends over time or `continuous data`.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.title('line plot Example')
plt.savefig('line_plot_day8.png')
plt.show()
```

#### ***Scatter Plots:***
Useful for exploring relationships between two `numerical variables`.

```python
plt.scatter(x, y)
plt.title('Scatter plot Example')
plt.savefig('scatter_plot-day8.png')
plt.show()
```

#### ***Bar Plots***:
Great for comparing `categorical data`.

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.savefig('Bar_plot_Day8.png')
plt.show()
```

#### ***Histograms:*** 
Used for displaying the distribution of a dataset.

```python
plt.hist(data)
plt.title('Histogram Example')
plt.savefig('Histogram-day8.png')
plt.show()
```

---

### 3. Customizing Plots:

Add `titles`, `axis labels`, `legends`, and `colors` to make your plots more informative and visually appealing.

```python
plt.plot(x, y, color='blue', marker='o', linestyle='--', linewidth=2, markersize=10)
plt.title('Trend over Time')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.legend(['Line'])
plt.savefig('Customized_plot_Day8.png')
plt.show()
```

## Exercises:
#### 1.Create a `Line Plot`:

Visualize a trend (e.g., sales growth over time) using a line plot.

```python
import pandas as pd

data = {
    'date': ['2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01', '2023-05-01'],
    'sales': [200, 250, 300, 350, 400]
}

df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])
```

```python
plt.plot(df['date'], df['sales'])
plt.title('Sales Growth Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

#### 2. Plot a Bar Chart:

Compare `categories` in your dataset (e.g., sales by product category).

```python
data = {
    'category': ['Electronics', 'Clothing', 'Groceries', 'Furniture', 'Toys'],
    'sales': [12000, 9000, 15000, 7000, 5000]
}

df = pd.DataFrame(data)
```

```python
plt.bar(df['category'], df['sales'])
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

#### 3. Customize a Plot:

Add `titles`, `axis labels`, and a `legend` to your plot.

```python
data = {
    'date': ['2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01', '2023-05-01'],
    'sales': [200, 250, 300, 350, 400]
}

df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])
```

```python
print(df)
plt.plot(df['date'], df['sales'], label='Sales')
plt.title('Sales Growth')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
