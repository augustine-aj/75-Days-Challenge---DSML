#### 2. Load the Dataset and Inspect its Structure:
Use `Pandas` to load the dataset and inspect its structure to understand the data.

```python
import pandas as pd

df = pd.read_csv("D:/75 Days Challenge/Day_6_bmi_dataset.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

#### 3. Clean the Dataset:
Handle `missing values`, `remove outliers`, or make other necessary data cleaning tasks to prepare the data for analysis.

```python
df.fillna(method='ffill', inplace=True)

df = df[df['Height'] < df['Height'].quantile(0.99)]
```

```python
df.head()
```

#### 4. Analyze the Dataset:
Use `Pandas` `grouping`, `filtering`, and `sorting` functions to explore the dataset further.

```python
grouped_data = df.groupby('Gender').mean()
```

```python
print("Grouped Data by Gender (Mean):\n")
grouped_data
```

```python
filtered_data = df[df['Weight'] > 100]
```

```python
print("\nFiltered Data (Weight > 100):\n")
filtered_data
```

```python
sorted_data = df.sort_values(by='Weight', ascending=False)
```

```python
print("\nSorted Data by Weight (Descending):\n")
sorted_data.head()
```

#### 5. Visualize Key Relationships and Trends:
Use `Matplotlib` or `Seaborn` to create visualizations and analyze key relationships within the data.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x='Height', y='Weight', data=df)
plt.title('Relationship between Height and Weight')
plt.savefig('Scatter_Day-10.png')
plt.show()

sns.barplot(x='Gender', y='Height', data=df)
plt.title('Category-wise Comparison')
plt.savefig('Bar_Day-10.png')
plt.show()

sns.histplot(df['Weight'], bins=20, kde=True)
plt.title('Distribution of Weight')
plt.savefig('Hist_Day-10.png')
plt.show()
```

#### 6.Present Insights Based on the Analysis:
Create a few summary plots and provide insights from the data analysis.

```python
plt.plot(df.index, df['Height'])  # Plot height by row index (as a proxy for trends)
plt.title('Height Trend Over Rows')
plt.xlabel('Row Index')
plt.ylabel('Height')
plt.show()

correlation = df[['Height', 'Weight', 'Index']].corr()
sns.heatmap(correlation, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
