## Theory Overview

#### Probability Basics:
- **Random Variables**: Variables that can take on different values based on the outcome of a random phenomenon.
- **Events**: Possible outcomes or combinations of outcomes.
- **Probability Rules**: Key rules include the **addition rule** (for mutually exclusive events) and **multiplication rule** (for independent events).

---

#### Discrete vs. Continuous Distributions:
- **Discrete Distribution**: Deals with countable outcomes (e.g., Binomial distribution).
- **Continuous Distribution**: Deals with an infinite range of possible outcomes (e.g., Normal distribution).

## Practical Exercise in Python
### 1. Calculating Basic Probabilities
Using Python, let's calculate a few probabilities.

```python
# Basic probability calculation
# Probability of getting a 3 or 5 when rolling a fair 6-sided die

# Define the event space and the sample space
sample_space = {1, 2, 3, 4, 5, 6}
event_space = {3, 5}

# Calculate probability
probability = len(event_space) / len(sample_space)
print("Probability of rolling a 3 or 5:", probability)
```

### 2. Working with Discrete Distributions: Binomial Distribution
Let's simulate a binomial distribution, where we calculate the probability of getting a certain number of successes in multiple trials.

For example, imagine flipping a coin 10 times and wanting to know the probability of getting exactly 6 heads.

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10  # number of trials (e.g., coin flips)
p = 0.5  # probability of success (e.g., probability of getting heads)

# Generate binomial distribution
x = np.arange(0, n+1)
binomial_probabilities = binom.pmf(x, n, p)

# Plotting
plt.bar(x, binomial_probabilities, color='skyblue')
plt.xlabel("Number of Successes (Heads)")
plt.ylabel("Probability")
plt.title("Binomial Distribution: Probability of Heads in 10 Coin Flips")
plt.show()
```

This code produces a bar chart showing the probability distribution of getting 0 to 10 heads in 10 coin flips.

### 3. Working with Continuous Distributions: Normal Distribution
Next, letâ€™s plot a normal distribution to see how continuous distributions are visualized.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Parameters for the normal distribution
mean = 0    # Mean of the distribution
std_dev = 1 # Standard deviation of the distribution

# Generate values for the x-axis
x_values = np.linspace(-4, 4, 1000)
# Get the corresponding normal distribution values (probability density)
normal_distribution = norm.pdf(x_values, mean, std_dev)

# Plotting
plt.plot(x_values, normal_distribution, color='purple')
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Normal Distribution (mean=0, std_dev=1)")
plt.grid()
plt.show()
```

The code above produces a smooth bell-shaped curve, representing a normal distribution with mean = 0 and standard deviation = 1.
