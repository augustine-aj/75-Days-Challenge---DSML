# Percentiles, Quartiles, and Data Visualization

In today’s session, I covered important statistical concepts and learned how to visualize data effectively using `Matplotlib` and `Seaborn`.

## Theory:

### Percentiles:
Percentiles are values that divide a dataset into 100 equal parts. The nth percentile is the value below which n% of the data falls. For example, the 25th percentile (Q1) is the value below which 25% of the data lies.

**Formula:**

$$
P_n = \left( \frac{n}{100} \right) \times (N + 1)
$$

Where:
- \( N \) is the number of data points
- \( n \) is the percentile.

### Quartiles:
Quartiles divide data into four equal parts:
- **Q1 (First Quartile)** – 25% of the data lies below this value.
- **Q2 (Second Quartile)** – This is the median of the dataset, splitting it into two halves.
- **Q3 (Third Quartile)** – 75% of the data lies below this value.

The **Interquartile Range (IQR)** is the range between Q3 and Q1, representing the middle 50% of the data. The formula for IQR is:

$$
\text{IQR} = Q3 - Q1
$$

IQR helps identify outliers (values outside the range \( Q1 - 1.5 \times \text{IQR} \) and \( Q3 +1.5 \times \text{IQR} \)).

## Data Visualization Basics:

### Histograms:
Histograms help visualize the frequency distribution of a dataset. They display the number of observations within each bin (range of values). Histograms are useful for understanding the shape of the data distribution.

### Box Plots:
Box plots (or box-and-whisker plots) provide a summary of a dataset’s distribution, showing the minimum, first quartile (Q1), median, third quartile (Q3), and maximum values. They are an effective way to identify outliers and understand the spread of the data.

### Scatter Plots:
Scatter plots show the relationship between two variables. Each point represents an observation in the dataset. Scatter plots are useful for identifying trends, clusters, and outliers.

### Calculating Percentiles and Quartiles:
I used **NumPy** and **Pandas** to calculate percentiles and quartiles in Python, which helped me better understand the data distribution and how to identify outliers.


### Sample Data:

```python
import numpy as np
import pandas as pd

data = [55, 89, 72, 45, 68, 78, 99, 83, 67, 92, 75, 85, 63, 61, 88, 77, 60, 71, 80, 95]

scores = pd.Series(data)
```

#### 1. Calculating Percentiles and Quartiles:
Percentiles and Quartiles can be calculated using NumPy and Pandas:

```python
percentile_25 = np.percentile(scores, 25)  # 25th percentile (Q1)
percentile_50 = np.percentile(scores, 50)  # 50th percentile (Median)
percentile_75 = np.percentile(scores, 75)  # 75th percentile (Q3)

IQR = percentile_75 - percentile_25

print(f"25th Percentile (Q1): {percentile_25}")
print(f"50th Percentile (Median): {percentile_50}")
print(f"75th Percentile (Q3): {percentile_75}")
print(f"IQR (Interquartile Range): {IQR}")
```

### 2. Visualizing Data with Matplotlib and Seaborn:
#### Histograms:
A histogram helps to understand the distribution of values in your dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(scores, kde=True, bins=10, color='blue', alpha=0.7)
plt.title('Histogram of Exam Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.show()
```

#### Box Plot:
A box plot shows the spread of data and helps to identify outliers.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=scores, color='green')
plt.title('Box Plot of Exam Scores')
plt.xlabel('Scores')
plt.show()
```

#### Scatter Plot:
A scatter plot is helpful when exploring the relationship between two variables. For this example, let's create two related variables.

```python
hours_studied = [3, 5, 2, 6, 4, 7, 9, 5, 3, 8, 4, 6, 5, 7, 6, 7, 5, 4, 6, 9]

df = pd.DataFrame({'Hours Studied': hours_studied, 'Scores': scores})

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Hours Studied', y='Scores', data=df, color='purple')
plt.title('Scatter Plot: Hours Studied vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.show()
```
