## Theory: Normal Distribution, Z-scores, and Central Limit Theorem (CLT)

### 1. Normal Distribution
The normal distribution is a bell-shaped, symmetric distribution that describes how data points are spread around the mean in many natural and human phenomena. It has these key characteristics:

- The **mean**, **median**, and **mode** are all at the center of the distribution.
- Data is **symmetrically distributed** around the mean.
- The **68-95-99.7 rule** states that:
  - 68% of data falls within 1 standard deviation from the mean.
  - 95% falls within 2 standard deviations.
  - 99.7% falls within 3 standard devia#tions.

## 2. Z-Scores
A **Z-score** is a measure that tells us how many standard deviations a data point is from the mean. It helps standardize different datasets, making them comparable. For example, if a data point has a Z-score of 2, it is 2 standard deviations above t#he mean.

### Formula$$
Z = \frac{X - \mu}{\sigma}
$$

where:
- \( X \) is the data point
- \( \mu \) is the mean
- \( \sigma \) is the standard deviati#on
s the standard deviation

## 3. Central Limit Theorem (CLT)
The **Central Limit Theorem** states that if we take a large enough number of random samples from any population (regardless of the populationâ€™s original distribution), the distribution of the sample means will tend to be normal. This holds true even if the original data isn't normally distributed. CLT is crucial because it allows us to:

- **Make inferences** about population parameters using sample data.
- **Approximate a sample mean distribution** as normal, which is essential for many statistical tests and machine learning algorithms.

## Practical: Implementing These Concepts in Python
Hereâ€™s how you can apply these concepts in Python with practical examples:

### 1. Normal Distribution and Z-scores
We'll generate a normally distributed dataset, calculate Z-scores for data points, and visualize the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

data = np.random.normal(loc=0, scale=1, size=1000)

sns.histplot(data, kde=True)
plt.title("Normal Distribution of Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

z_scores = (data - np.mean(data)) / np.std(data)
print("Sample Z-Scores:", z_scores[:5])
```

### Explanation:

- We generate a dataset that follows a normal distribution using np.random.normal().
- Using Seaborn, we plot a histogram with a kernel density estimate (KDE) to show the distribution curve.
- Z-scores are calculated for each data point, which shows how far each point is from the mean in terms of standard deviations.
