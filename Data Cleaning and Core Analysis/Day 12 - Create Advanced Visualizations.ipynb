## Tasks:
### 1. Visual Exploration:
### Visualize Key Relationships:

- Use `scatter plots`, `bar plots`, and `box plots` to investigate relationships between key variables.
- Start with basic plots and move to more advanced visualizations, such as joint plots or facet grids, to explore bivariate or multivariate relationships.
#### Advanced Visualization Code:

```python
import pandas as pd

df = pd.read_csv("D:/75 Days Challenge/Day_6_bmi_dataset.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lmplot(x='Height', y='Weight', data=df, hue='Gender', height=6, aspect=1.5)
plt.title("Height vs Weight with Regression Line")
plt.show()
```

```python
sns.boxplot(x='Gender', y='Weight', data=df, palette='Set3')
plt.title("Weight Distribution by Gender")
plt.show()
```

```python
sns.pairplot(df, hue='Gender', markers=["o", "s"], palette="husl")
plt.title("Pairplot for Multivariate Data Exploration")
plt.show()
```

```python
df_corr = df[['Height', 'Weight', 'Index']]
corr = df_corr.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

### 2. Design Visual Aesthetics:
### Enhance Plot Aesthetics:

Customize your plots with `titles`, `labels`, `legends`, and `color schemes` to make them visually appealing.
Use `Seaborn's` built-in themes or `Matplotlib` customization options to create `professional-looking plots`.
#### Advanced Aesthetics Code:

```python
sns.set(style="whitegrid")

plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Gender', y='Weight', data=df, palette='Blues_d')
ax.set_title("Weight Comparison by Gender", fontsize=16)
ax.set_xlabel("Gender", fontsize=12)
ax.set_ylabel("Weight (kg)", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0, linewidths=1, linecolor='black')
plt.title("Correlation Matrix with Aesthetic Improvements")
plt.show()
```

### 3. Advanced Interactive Visualizations:
Explore `interactive visualizations` using libraries like Plotly, which provide dynamic visuals that allow users to hover over data points for additional insights. This can help in presentations or reports where viewers need more interaction with the data.

#### Interactive Visualization Code:

```python
import plotly.express as px

fig = px.scatter(df, x='Height', y='Weight', color='Gender', title='Height vs Weight Interactive Scatter Plot')
fig.show()

# Interactive Heatmap
fig = px.imshow(corr, text_auto=True, color_continuous_scale='Viridis')
fig.update_layout(title='Interactive Correlation Heatmap')
fig.show()
```
