# Unsupervised Learning - Clustering Basics

## What is Clustering?
Clustering is an unsupervised learning technique used to group similar data points together without any labeled outputs.  
The goal is to identify underlying patterns or structures in data by organizing it into clusters based on feature similarity.

---

## K-Means Clustering

### How It Works:

1. **Initialization**:
   - Choose the number of clusters (\(k\)) to group the data into.
   - Randomly initialize \(k\) centroids (center points for each cluster).

2. **Assigning Points to Clusters**:
   - For each data point, calculate the distance to each centroid.
   - Assign the data point to the cluster whose centroid is closest.

3. **Updating Centroids**:
   - Recalculate the centroids of the clusters based on the mean of all points assigned to each cluster.

4. **Repeat**:
   - Reassign points and update centroids until centroids no longer change significantly or a predefined number of iterations is reached.

---

### Applications:
- **Customer segmentation** (e.g., grouping users by purchasing behavior).  
- **Image compression**.  
- **Anomaly detection**.  

---

## Visualizing Clusters
After clustering, it's common to visualize the results:  
- For **2D datasets**, plot the data points with different colors for each cluster.  
- Include **cluster centroids** for better understanding.

---

## Practical Implementation in Python

### 1. Perform K-Means Clustering:
- Use Scikit-learn's `KMeans` class to perform clustering.  
- Define the number of clusters (`n_clusters`).

### 2. Visualize the Clusters:
- Use Matplotlib to scatter-plot the data points and color them based on their cluster assignments.

---

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)

kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50)
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

## Key Insights from the Code

### Input Dataset
- The data points are randomly generated using `make_blobs` for simplicity.

### K-Means Algorithm
- The algorithm identifies 4 clusters in the data and assigns each point to the nearest cluster.

## Visualization
- Each cluster is shown in a distinct color.
- The centroids of the clusters are marked with a red 'X'.

### Key Parameters of K-Means
1. **`n_clusters`**:
   - Specifies the number of clusters to form.
   - Example: `n_clusters=4` creates 4 clusters.

2. **`init`**:
   - Determines the method for initializing centroids.
   - Example: `k-means++` ensures smarter and more efficient centroid initialization.

3. **`max_iter`**:
   - Sets the maximum number of iterations for the algorithm to converge.
   - Default: 300 iterations.
