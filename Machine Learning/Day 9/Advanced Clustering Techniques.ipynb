# Advanced Clustering Techniques

In this session, we explore advanced unsupervised learning methods, focusing on **hierarchical clustering** and **DBSCAN** (Density-Based Spatial Clustering of Applications with Noise). These techniques are effective for datasets with no predefined number of clusters or with noisy data.

---

## 1. Hierarchical Clustering

Hierarchical clustering builds a **tree-like structure** of clusters, offering flexibility in analyzing different levels of granularity. It can be performed in two ways:

### What is Hierarchical Clustering?
- **Agglomerative (Bottom-Up)**: 
  - Start with each data point as its own cluster.
  - Merge clusters iteratively until all points form one cluster.
- **Divisive (Top-Down)**: 
  - Start with all points in one cluster.
  - Split the cluster iteratively into smaller clusters.

### Key Concepts:
1. **Dendrogram**:
   - A tree diagram that visually represents the process of merging or splitting clusters.
   - The height of each merge represents the distance (or dissimilarity) between clusters.

2. **Linkage Criteria**:
   - Determines how the distance between clusters is calculated:
     - **Single Linkage**: Minimum distance between points in two clusters.
     - **Complete Linkage**: Maximum distance between points in two clusters.
     - **Average Linkage**: Average distance between points in two clusters.

### Advantages:
- Does not require specifying the number of clusters beforehand.
- Suitable for small to medium-sized datasets.

### Disadvantages:
- Computationally expensive for large datasets.
- Sensitive to noise and outliers.

---

### Code Example: Hierarchical Clustering

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

X, y = make_blobs(n_samples=200, centers=3, cluster_std=1.05, random_state=42)

plt.figure(figsize=(10, 7))
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.show()

hc = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_hc, cmap='viridis')
plt.title('Hierarchical Clustering')
plt.show()
```

## 2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
### What is it?
DBSCAN groups points that are close to each other based on density and marks points in low-density regions as noise.

### Key Parameters:
- **Epsilon (eps):** Maximum distance for two points to be considered neighbors.
- **Min Samples:** Minimum number of points required to form a dense region.
## Advantages:
- Can identify clusters of arbitrary shapes.
- Handles outliers (noise) well.

## Code Example: DBSCAN

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import numpy as np

X, y = make_moons(n_samples=300, noise=0.05, random_state=42)

dbscan = DBSCAN(eps=0.2, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='viridis', s=50)
plt.title('DBSCAN Clustering')
plt.show()
```

## When to Use Each Clustering Technique?

| **Clustering Method** | **When to Use**                                                                 |
|------------------------|---------------------------------------------------------------------------------|
| **Hierarchical**       | - If you want a visual representation of clusters (dendrogram).               |
|                        | - If you're unsure about the number of clusters in the dataset.               |
| **DBSCAN**             | - For datasets with arbitrary-shaped clusters.                                |
|                        | - When the dataset contains noise or outliers.                                |


## Takeaways:
1. **Hierarchical Clustering** is great for a hierarchical view of clusters but may struggle with large datasets.
2. **DBSCAN** excels at handling noise and detecting non-linear clusters but requires careful tuning of parameters like `eps` and `min_samples`.

    
Both methods are valuable tools for unsupervised learning and help uncover hidden patterns in your data.
